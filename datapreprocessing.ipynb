```python
import pandas as pd
import numpy as np
import pickle 
```

```python
df = pd.read_csv("t20_wc.csv")
```

```python
df.isnull().sum()
```

```python
df[df["city"].isnull()]["venue"].value_counts()
```

```python
df["city"]=df["city"].mask(df["city"].isnull(),df["venue"].str.split().str.get(0))
```

```python
df.isnull().sum()
```

```python
df.drop(columns = ["venue"],inplace =True)
```

   #data filtering 
* each city should have a record of al least 5 inngs played

```python
df["city"].unique()
```

```python
(20*6)*5
```

```python
eligible_cities = df ["city"].value_counts()
```

```python
eligible_cities
```

```python
# No ball wide ball extras ball we will take max balls bowled  in the cities
```

```python
eligible_cities = eligible_cities [eligible_cities > 600].index.tolist()
print(eligible_cities)
```

```python
df=df[df["city"].isin(eligible_cities)]
```

```python
df
```

```python
# for current score 
```

```python
df["current_score"]=df.groupby("match_id")["runs"].cumsum()
```

```python
df["over"]=df["ball"].astype(int)
```

```python
df["ball_no"]=df["ball"].astype(str).str.extract(r"\d.(\d)").astype(int)
```

```python
df["ball_no"]=df["ball"].astype(str).str.extract(r"\d.(\d)").astype(int)
```

```python
df["total_delivers"]=(df["over"]*6) + df["ball_no"]
```

```python
df["ball_left"] = 120 - df["total_delivers"]
```

```python
df["ball_left"].mask(df["ball_left"] < 0 , 0 , inplace = True)
```

```python
df["player_dismissed"]=df["player_dismissed"].mask(df["player_dismissed"] != "0" , 1 ).astype(int)
```

```python
df["player_dismissed"] = df.groupby("match_id")["player_dismissed"].cumsum()
```

```python
df
```

# WICKET LEFT

```python
df["wicket_left"] = 10 - df["player_dismissed"]
```

# FOR CRR

```python
df["crr"] = (df["current_score"]*6/df["total_delivers"])
df
```

# FOR LAST 5 OVER

```python
df["last_five"]=df.groupby("match_id")["runs"].rolling(window = 30).sum().values.tolist()
```

# FOR FINAL SCORE 

```python
total_runs_table = df.groupby("match_id")["runs"].sum().reset_index()
```

```python
total_runs_table
```

```python
df = df.merge(total_runs_table, on  = "match_id")
```

```python
df
```

```python
df.columns
```

```python
final_df = \
df[['batting_team', 'bowling_team', 'current_score','ball_left','city','wicket_left', 'crr', 'last_five','runs_y']]
```

```python
final_df
```

```python
final_df.dropna(inplace = True)
```

```python
final_df.isnull().sum()
```

```python
final_df=final_df.sample(final_df.shape[0])
```

```python
final_df
```

```python
x = final_df.drop(columns = ["runs_y"])
y = final_df["runs_y"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,X_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 1)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
trf = ColumnTransformer ([
    ('trf',OneHotEncoder (sparse_output = False,drop = 'first'),["batting_team","bowling_team","city"])
]
,remainder = 'passthrough') 
```

```python
pipe = Pipeline(steps = [
       ('step1' ,trf),
       ('step2' ,StandardScaler()),
       ('step3' ,XGBRegressor(n_estimators = 1000 ,learning_rate = 0.2 , max_depth = 12, random_state = 1))])
```

```python
pipe.fit(x_train,y_train)
y_pred = pipe.predict(X_test)
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```

```python
pickle.dump(pipe,open('pipe.pkl','wb'))
```
